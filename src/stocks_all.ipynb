```python
import datetime
from datetime import timedelta
def compute_growth(GrowthFinalPeriod = parser.parse(right_now), 
                   GrowthInitPeriod = datetime.datetime(2017, 3, 1, 0, 0)):
    growth = {}
    price = {}
    print('computing the gain from:', GrowthInitPeriod, 'to:', GrowthFinalPeriod)
    for k in alldata.keys():
        nextFlag = False
        finaldate = GrowthFinalPeriod
        while finaldate not in alldata[k]:
            finaldate = finaldate - timedelta(days=1)
            if finaldate < GrowthFinalPeriod - timedelta(days=7):
                nextFlag = True
                break
        firstdate = GrowthInitPeriod
        while firstdate not in alldata[k]:
            firstdate = firstdate - timedelta(days=1)
            if firstdate < GrowthInitPeriod - timedelta(days=7):
                nextFlag = True
                break
        if nextFlag:
            continue
        last_year_relative = (alldata[k][finaldate][-1] -  alldata[k][firstdate][-1]) / alldata[k][firstdate][-1]
        price[k] = alldata[k][finaldate][-1]
        growth[k] = last_year_relative
    sumavg = 0
    for k in growth:
        sumavg += growth[k]
    print('average of all 500 stocks is:', str(sumavg/len(growth.keys()))[0:5], ' over:', len(growth.keys()), ' stocks')
    print('\n')
    return growth, price

def compute_volatility(VolFinalPeriod = parser.parse(right_now), 
                   VolInitPeriod = datetime.datetime(2017, 3, 1, 0, 0), start_money=1000):
    fluctuations = {}
    upjumps = {}
    price = {}
    total_simple_strategy = {}
    total_growth_strategy = {}
    print('computing the volatility from:', VolInitPeriod, 'to:', VolFinalPeriod)
    for k in alldata.keys():
        nextFlag = False
        finaldate = VolFinalPeriod
        while finaldate not in alldata[k]:
            finaldate = finaldate - timedelta(days=1)
            if finaldate < VolFinalPeriod - timedelta(days=7):
                nextFlag = True
                break
        firstdate = VolInitPeriod
        while firstdate not in alldata[k]:
            firstdate = firstdate - timedelta(days=1)
            if firstdate < VolInitPeriod - timedelta(days=7):
                nextFlag = True
                break
        if nextFlag:
            continue
        absolute_sum_fluctuations = 0
        absolute_sum_ups = 0
        #simple_strategy_gain = 0
        last_price = None
        num_days = 0
        total_left = start_money
        init_buy_cnt = 0
        init_buy_leftover = 0
        holding = 0
        for date_i in sorted(alldata[k]):
            if date_i < VolInitPeriod or date_i > VolFinalPeriod:
                continue
            if last_price is not None:
                absolute_sum_fluctuations += abs(alldata[k][date_i][-1] - last_price)/last_price
                absolute_sum_ups += abs(alldata[k][date_i][-1] - last_price)/last_price if alldata[k][date_i][-1]>last_price else 0
                if alldata[k][date_i][-1] > last_price and holding > 0:
                    total_left += holding * alldata[k][date_i][-1]
                    holding = 0
                if alldata[k][date_i][-1] < last_price and total_left > alldata[k][date_i][-1]:
                    cnt = total_left // alldata[k][date_i][-1]
                    total_left -= cnt * alldata[k][date_i][-1]
                    holding += cnt
                num_days += 1
            else:
                init_buy_cnt = (total_left // alldata[k][date_i][-1])
                init_buy_leftover = total_left - (alldata[k][date_i][-1] * init_buy_cnt)
            last_price = alldata[k][date_i][-1]
        total_left += holding * alldata[k][finaldate][-1]
        fluctuations[k] = absolute_sum_fluctuations/num_days
        upjumps[k] = absolute_sum_ups/num_days
        price[k] = alldata[k][finaldate][-1]
        total_simple_strategy[k] = total_left
        total_growth_strategy[k] = alldata[k][finaldate][-1]*init_buy_cnt + init_buy_leftover
    return fluctuations, upjumps, total_simple_strategy,total_growth_strategy, price


```

```python
start_money=1000
fluctuations, upjumps, total_simple_strategy, total_growth_strategy, price = compute_volatility(
    VolFinalPeriod=datetime.datetime(2018, 5, 11, 0, 0), 
    VolInitPeriod=datetime.datetime(2018, 4, 1, 0, 0), 
    start_money=start_money) #parser.parse(right_now) - timedelta(365)) #
for w in sorted(total_simple_strategy, key=total_simple_strategy.get, reverse=True):
    wkey = w.split('.')[0]
    print(wkey, '-(with $'+str(start_money)+' start) strategy:$'+str(total_simple_strategy[w])[0:5] + ' growth:$' + str(total_growth_strategy[w])[0:5] + ' fin_price:$' + str(price[w]).split('.')[0], '--'+ desc[wkey].replace('\t',';') if wkey in desc else wkey)
```

```python
import pandas as pd
import datetime
import numpy as np
from pandas.tseries.holiday import USFederalHolidayCalendar
cal = USFederalHolidayCalendar()
holidays = cal.holidays(start='2014-01-01', end='2018-12-31').to_pydatetime()
from datetime import timedelta

weatherdata = pd.read_csv('../data/1306639.csv')
print(weatherdata.columns)
weather_dict = {}
for ix in range(len(weatherdata)):
    weather_dict[parser.parse(weatherdata.iloc[ix][2])] = np.array(weatherdata.iloc[ix][3:])
```

```python
prev_price = alldata['NVDA.csv'][datetime.datetime(2016,1,4,0,0)][-1] 
prev_weather = weather_dict[datetime.datetime(2016,1,4,0,0)]
dataX = np.zeros((len(alldata['NVDA.csv']), 29), dtype=float)
dataY = np.zeros((len(alldata['NVDA.csv'])), dtype=float)
ix = 0
for day in sorted(weather_dict):
    if day in alldata['NVDA.csv']:
        price_jump = (alldata['NVDA.csv'][day][-1] - prev_price)/prev_price
        dataY[ix] = price_jump
        dataX[ix,0:9] = np.array(day.timetuple())
        dataX[ix,9] = ((day - timedelta(3)) in holidays)
        dataX[ix,10] = ((day - timedelta(4)) in holidays)
        dataX[ix,11:] = (weather_dict[day] - prev_weather)/prev_weather
        prev_price = alldata['NVDA.csv'][day][-1] 
        prev_weather = dataX[ix,11:] = weather_dict[day]
        ix += 1

#dataX = dataX[:,1:]
tempXminT = dataX[:,11+7]
tempXmaxT = dataX[:,11+6]
tempXrain = dataX[:,11+2]
dayofweekX = dataX[:,6]

```

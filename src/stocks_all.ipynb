```python
import main
import importlib as imp
imp.reload(main)
from dateutil import parser
```

```python
alldata = main.load_data('./prices/', datefrom='1-jan-2016', dateto=right_now)
info = [l.strip() for l in open('stklist_info.txt', 'r').readlines()]
infokeys= [i.split('\t')[0] for i in info]
desc = {};
for ix, k in enumerate(infokeys):
    desc[k] = info[ix]
```

```python
import datetime
from datetime import timedelta
def compute_growth(GrowthFinalPeriod = parser.parse(right_now), 
                   GrowthInitPeriod = datetime.datetime(2017, 1, 1, 0, 0)):
    growth = {}
    price = {}
    print('computing the gain from:', GrowthInitPeriod, 'to:', GrowthFinalPeriod)
    for k in alldata.keys():
        nextFlag = False
        finaldate = GrowthFinalPeriod
        while finaldate not in alldata[k]:
            finaldate = finaldate - timedelta(days=1)
            if finaldate < GrowthFinalPeriod - timedelta(days=7):
                nextFlag = True
                break
        firstdate = GrowthInitPeriod
        while firstdate not in alldata[k]:
            firstdate = firstdate - timedelta(days=1)
            if firstdate < GrowthInitPeriod - timedelta(days=7):
                nextFlag = True
                break
        if nextFlag:
            continue
        last_year_relative = (alldata[k][finaldate][-1] -  alldata[k][firstdate][-1]) / alldata[k][firstdate][-1]
        price[k] = alldata[k][finaldate][-1]
        growth[k] = last_year_relative
    sumavg = 0
    for k in growth:
        sumavg += growth[k]
    print('average of all 500 stocks is:', str(sumavg/len(growth.keys()))[0:5], ' over:', len(growth.keys()), ' stocks')
    print('\n')
    return growth, price

```

```python
g, p = compute_growth(parser.parse(right_now), datetime.datetime(2018, 2, 28, 0, 0))
for w in sorted(g, key=g.get, reverse=True):
    wkey = w.split('.')[0]
    print(wkey, '- growth: '+str(g[w])[0:5], 'price: $' + str(p[w]).split('.')[0], '--'+ desc[wkey].replace('\t',';') if wkey in desc else wkey)
```
